```python
# Para habilitar método str.get.dummies() y para restablecer orden numérico de los strings:
for i in range(1,19):
    if i<10: 
        df_doct['CCAARESI'].replace({i:'0'+str(i)},inplace=True)
    else:
        df_doct['CCAARESI'].replace({i:str(i)},inplace=True)
    
df_doct['CCAARESI'].sort_values().unique()
```

#### CCAA de residencia: 
'01'=Andalucía, '02'=Aragón, '03'=Asturias (Principado de), '04'=Balears (Illes), '05'=Canarias, '06'=Cantabria, '07'=Castilla y León, '08'=Castilla - La Mancha, '09'=Cataluña, '10'=Comunidad Valenciana, '11'=Extremadura, '12'=Galicia, '13'=Madrid (Comunidad de), '14'=Murcia (Región de), '15'=Navarra (Comunidad Foral de), '16'=País Vasco, '17'=Rioja (La), '18'=Ceuta y Melilla.

```python
# Nueva variable 'edad09': edad en el momento de la encuesta 
df_doct['edad09']=2009-df_doct['ANONAC']
df_doct['edad09'].describe()
```

```python
# Nueva variable 'fem': 0=Hombre, 1=Mujer.
df_doct['fem']=df_doct['SEXO'].mask(df_doct['SEXO']==1,0)
df_doct['fem']=df_doct['fem'].mask(df_doct['SEXO']==2,1)
df_doct['fem']=df_doct['fem'].astype('int')
df_doct['fem'].describe()
```

```python
D_ccaa={'01':'Andalucía','02':'Aragón','03':'Asturias',
                        '04':'Baleares','05':'Canarias','06':'Cantabria','07':'Castilla y León',
                        '08':'Castilla-La Mancha','09':'Cataluña','10':'Comunidad Valenciana','11':'Extremadura',
                        '12':'Galicia','13':'Madrid','14':'Murcia','15':'Navarra','16':'País Vasco',
                        '17':'La Rioja','18':'Ceuta y Melilla'}

ax=sns.displot(df_doct.sort_values(by='CCAARESI'),x='CCAARESI',bins=18,hue='fem',multiple='dodge',
               stat='density',discrete=True,palette='Paired',legend=False)
ax.set(title='Distribución de doctores por CCAA y sexo',xlabel='CCAA',ylabel='Frecuencia relativa')
ax.set_xticklabels(rotation = 90)
ax.set(xticklabels=D_ccaa.values())
plt.legend(title='Sexo',loc='center right',labels=['Mujer','Hombre'],frameon=False)
```

```python
# Nueva variable 'esp': 1=Nacionalidad española, 0=Nacionalidad no española
df_doct['esp']=df_doct['RELA'].mask((df_doct['RELA']==1)|(df_doct['RELA']==2),1)
df_doct['esp']=df_doct['esp'].mask((df_doct['RELA']!=1)&(df_doct['RELA']!=2),0)
df_doct['esp']=df_doct['esp'].astype('int')
df_doct['esp'].describe()
```

```python
df_doct[['esp','RELA']].groupby('RELA').agg(['unique'])
```

```python
# Nueva variable 'eu27': 1=Nacido en EU27, 0=No nacido en EU27
df_doct['eu27']=df_doct['CONTNACIM'].mask(df_doct['CONTNACIM']=='EU27',1)
df_doct['eu27']=df_doct['eu27'].mask(df_doct['CONTNACIM']!='EU27',0)
df_doct['eu27']=df_doct['eu27'].astype('int')
df_doct['eu27'].describe()
```

```python
# Nuevas variables binarias para continente de nacimiento: 
df_doct[['afr','asia','eu27','eeuuca','ocea','rest_am','rest_eu']]=df_doct['CONTNACIM'].str.get_dummies()
df_doct[['afr','asia','eu27','eeuuca','ocea','rest_am','rest_eu','CONTNACIM']].groupby('CONTNACIM').agg(['unique'])
```

```python
df_doct[['afr','asia','eu27','eeuuca','ocea','rest_am','rest_eu']].describe()
```

```python
# Nuevas variables binarias para primera nacionalidad: 
df_doct[['na_afr','na_asia','na_eu27','na_eeuuca'
         ,'na_rest_am','na_rest_eu']]=df_doct['CONTNAC1'].str.get_dummies()
df_doct[['na_afr','na_asia','na_eu27','na_eeuuca',
         'na_rest_am','na_rest_eu','CONTNAC1']].groupby('CONTNAC1').agg(['unique'])
```

**'CODIDOC'**

Código de estudios de doctorado:

CIENCIAS NATURALES: 101=Matemáticas, 102=Informática y tecnologías de la información, 103=Ciencias físicas, 104=Ciencias químicas, 105=Ciencias de la tierra y medio ambiente, 106=Biología (excluyendo agricultura y ciencias médicas), 107=Otras ciencias naturales

INGENIERÍA Y TECNOLOGÍA: 201=Ingeniería civil, 202=Ingeniería eléctrica, electrónica y de telecomunicaciones, 203=Ingeniería mecánica, 204=Ingeniería química, 205=Ingeniería de materiales, 206=Ingeniería médica, 207=Ingeniería medioambiental, 208=Biotecnología medioambiental, 209=Biotecnología industrial, 210=Nanotecnología, 211=Otras ingenierías y tecnologías (comida, bebida y otras)

CIENCIAS MÉDICAS: 301=Medicina básica, 302=Medicina clínica, 303=Ciencias de la salud, 304=Biotecnología médica, 305=Otras ciencias médicas (forenses y otras ciencias médicas)

CIENCIAS DE LA AGRICULTURA: 401=Agricultura, ciencias forestales y piscifactorías, 402=Ciencias de los animales y de la leche, 403=Veterinaria, 404=Biotecnología agrícola, 405=Otras ciencias de la agricultura

CIENCIAS SOCIALES: 501=Psicología, 502=Economía y empresas, 503=Ciencias de la educación, 504=Sociología, 505=Derecho, 506=Ciencias políticas, 507=Geografía económica y social, 508=Periodismo y comunicaciones, 509=Otras ciencias sociales

HUMANIDADES: 601=Historia y arqueología, 602=Lenguaje y literatura, 603=Filosofía, ética y religión, 604=Arte (historia del arte, bellas artes y música), 605=Otras humanidades


```python
# Nueva variable 'codidoc': familia de estudios de doctorado: 
df_doct['codidoc']=df_doct['CODIDOC'].apply(lambda x:str(int(x/100)))
```

```python
# Nuevas variables binarias para estudios de doctorado: 
df_doct[['cc_naturales','ingeniería y tecn','cc_médicas','cc_agricultura','cc_sociales',
        'humanidades']]=df_doct['codidoc'].str.get_dummies()
df_doct[['cc_naturales','ingeniería y tecn','cc_médicas','cc_agricultura','cc_sociales',
        'humanidades','CODIDOC']].groupby('CODIDOC').agg(['unique'])
```

**'FINADOC'¶**

Fuente principal de financiación durante los estudios de doctorado: 1=Beca de la Administración Pública (central, autonómica, local,etc.), 2=Beca de otro tipo de institución española, 3=Beca de una Administración Pública extranjera, 4=Beca de otro tipo de institución extranjera, 5=Trabajó como profersor y/o ayudante de investigación, 6=Otra ocupación, 7=Subvención reembolsada por el empleador, 8=Préstamos, ahorros personales y/o ayuda familiar, 9=Otras formas.


```python
df_doct['fi_doc']=pd.cut(df_doct['FINADOC'],[0,4,5,7,8,9],
                         labels=['beca','docencia/investigación','trabajo no docente','patrimonio','otra'])
df_doct[['fi_doc','FINADOC']].groupby('FINADOC').agg(['unique'])
```

```python
df_doct[['beca','docente/investigador','otra financiación','patrimonio',
         'trabajo no docente']]=df_doct['fi_doc'].str.get_dummies()
```

```python
df_doct[['beca','docente/investigador','trabajo no docente','patrimonio',
         'otra financiación','fi_doc']].groupby('fi_doc').agg(['unique'])
```

```python
df_doct[['beca','docente/investigador','trabajo no docente','patrimonio',
         'otra financiación']].describe()
```

```python
ax=sns.displot(df_doct.sort_values(by='fi_doc'),x='fi_doc',bins=5,discrete=True,hue='fem',multiple='dodge',
               stat='density', palette='Paired',legend=False)
ax.set(title='Fuentes de financiación según sexo',xlabel='Fuente de financiación',ylabel='Densidad')
ax.set_xticklabels(['Beca','Docencia o\n investigación','trabajo no\n docente','Patrimonio','Otra'],rotation = 90)
plt.legend(title='Sexo',loc='center right',labels=['Mujer','Hombre'],frameon=False)
```
