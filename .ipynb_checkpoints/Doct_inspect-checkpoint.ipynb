```python
# Para habilitar mtodo str.get.dummies() y para restablecer orden numrico de los strings:
for i in range(1,19):
    if i<10: 
        df_doct['CCAARESI'].replace({i:'0'+str(i)},inplace=True)
    else:
        df_doct['CCAARESI'].replace({i:str(i)},inplace=True)
    
df_doct['CCAARESI'].sort_values().unique()
```

#### CCAA de residencia: 
'01'=Andaluca, '02'=Aragn, '03'=Asturias (Principado de), '04'=Balears (Illes), '05'=Canarias, '06'=Cantabria, '07'=Castilla y Len, '08'=Castilla - La Mancha, '09'=Catalua, '10'=Comunidad Valenciana, '11'=Extremadura, '12'=Galicia, '13'=Madrid (Comunidad de), '14'=Murcia (Regin de), '15'=Navarra (Comunidad Foral de), '16'=Pas Vasco, '17'=Rioja (La), '18'=Ceuta y Melilla.

```python
# Nueva variable 'edad09': edad en el momento de la encuesta 
df_doct['edad09']=2009-df_doct['ANONAC']
df_doct['edad09'].describe()
```

```python
# Nueva variable 'fem': 0=Hombre, 1=Mujer.
df_doct['fem']=df_doct['SEXO'].mask(df_doct['SEXO']==1,0)
df_doct['fem']=df_doct['fem'].mask(df_doct['SEXO']==2,1)
df_doct['fem']=df_doct['fem'].astype('int')
df_doct['fem'].describe()
```

```python
D_ccaa={'01':'Andaluca','02':'Aragn','03':'Asturias',
                        '04':'Baleares','05':'Canarias','06':'Cantabria','07':'Castilla y Len',
                        '08':'Castilla-La Mancha','09':'Catalua','10':'Comunidad Valenciana','11':'Extremadura',
                        '12':'Galicia','13':'Madrid','14':'Murcia','15':'Navarra','16':'Pas Vasco',
                        '17':'La Rioja','18':'Ceuta y Melilla'}

ax=sns.displot(df_doct.sort_values(by='CCAARESI'),x='CCAARESI',bins=18,hue='fem',multiple='dodge',
               stat='density',discrete=True,palette='Paired',legend=False)
ax.set(title='Distribucin de doctores por CCAA y sexo',xlabel='CCAA',ylabel='Frecuencia relativa')
ax.set_xticklabels(rotation = 90)
ax.set(xticklabels=D_ccaa.values())
plt.legend(title='Sexo',loc='center right',labels=['Mujer','Hombre'],frameon=False)
```

```python
# Nueva variable 'esp': 1=Nacionalidad espaola, 0=Nacionalidad no espaola
df_doct['esp']=df_doct['RELA'].mask((df_doct['RELA']==1)|(df_doct['RELA']==2),1)
df_doct['esp']=df_doct['esp'].mask((df_doct['RELA']!=1)&(df_doct['RELA']!=2),0)
df_doct['esp']=df_doct['esp'].astype('int')
df_doct['esp'].describe()
```

```python
df_doct[['esp','RELA']].groupby('RELA').agg(['unique'])
```

```python
# Nueva variable 'eu27': 1=Nacido en EU27, 0=No nacido en EU27
df_doct['eu27']=df_doct['CONTNACIM'].mask(df_doct['CONTNACIM']=='EU27',1)
df_doct['eu27']=df_doct['eu27'].mask(df_doct['CONTNACIM']!='EU27',0)
df_doct['eu27']=df_doct['eu27'].astype('int')
df_doct['eu27'].describe()
```

```python
# Nuevas variables binarias para continente de nacimiento: 
df_doct[['afr','asia','eu27','eeuuca','ocea','rest_am','rest_eu']]=df_doct['CONTNACIM'].str.get_dummies()
df_doct[['afr','asia','eu27','eeuuca','ocea','rest_am','rest_eu','CONTNACIM']].groupby('CONTNACIM').agg(['unique'])
```

```python
df_doct[['afr','asia','eu27','eeuuca','ocea','rest_am','rest_eu']].describe()
```

```python
# Nuevas variables binarias para primera nacionalidad: 
df_doct[['na_afr','na_asia','na_eu27','na_eeuuca'
         ,'na_rest_am','na_rest_eu']]=df_doct['CONTNAC1'].str.get_dummies()
df_doct[['na_afr','na_asia','na_eu27','na_eeuuca',
         'na_rest_am','na_rest_eu','CONTNAC1']].groupby('CONTNAC1').agg(['unique'])
```

```python
df_doct[['na_afr','na_asia','na_eu27','na_eeuuca','na_rest_am','na_rest_eu']].describe()
```

CIENCIAS DE LA AGRICULTURA: 401=Agricultura, ciencias forestales y piscifactoras, 402=Ciencias de los animales y de la leche, 403=Veterinaria, 404=Biotecnologa agrcola, 405=Otras ciencias de la agricultura

CIENCIAS SOCIALES: 501=Psicologa, 502=Economa y empresas, 503=Ciencias de la educacin, 504=Sociologa, 505=Derecho, 506=Ciencias polticas, 507=Geografa econmica y social, 508=Periodismo y comunicaciones, 509=Otras ciencias sociales

```python
# Nueva variable 'codidoc': familia de estudios de doctorado: 
df_doct['codidoc']=df_doct['CODIDOC'].apply(lambda x:str(int(x/100)))
```

```python
# Nuevas variables binarias para estudios de doctorado: 
df_doct[['cc_naturales','ingeniera y tecn','cc_mdicas','cc_agricultura','cc_sociales',
        'humanidades']]=df_doct['codidoc'].str.get_dummies()
df_doct[['cc_naturales','ingeniera y tecn','cc_mdicas','cc_agricultura','cc_sociales',
        'humanidades','CODIDOC']].groupby('CODIDOC').agg(['unique'])
```

**'FINADOC'**

Fuente principal de financiacin durante los estudios de doctorado: 1=Beca de la Administracin Pblica (central, autonmica, local,etc.), 2=Beca de otro tipo de institucin espaola, 3=Beca de una Administracin Pblica extranjera, 4=Beca de otro tipo de institucin extranjera, 5=Trabaj como profersor y/o ayudante de investigacin, 6=Otra ocupacin, 7=Subvencin reembolsada por el empleador, 8=Prstamos, ahorros personales y/o ayuda familiar, 9=Otras formas.


```python
df_doct['fi_doc']=pd.cut(df_doct['FINADOC'],[0,4,5,7,8,9],
                         labels=['beca','docencia/investigacin','trabajo no docente','patrimonio','otra'])
df_doct[['fi_doc','FINADOC']].groupby('FINADOC').agg(['unique'])
```

```python
df_doct[['beca','docente/investigador','otra financiacin','patrimonio',
         'trabajo no docente']]=df_doct['fi_doc'].str.get_dummies()
```
